```python
import Page1 as st
import pandas as pd
import plotly.express as px

st.set_page_config(
    page_title="HR Attrition Dashboard",
    layout="wide"
)

st.title("HR Attrition Dashboard")

# Đọc dữ liệu
# df = pd.read_csv("employee_data.csv")

# Kiểm tra các cột cần có
required_cols = ["Attrition", "Department", "JobRole"]
missing_cols = [col for col in required_cols if col not in df.columns]

if missing_cols:
    st.error(f"Thiếu cột: {missing_cols}")
    st.stop()

# =========================
# A. Tổng quan Attrition
# =========================
st.header("A. Tổng quan Attrition")

total_employees = len(df)
attrition_yes = df[df["Attrition"] == "Yes"].shape[0]
attrition_rate = attrition_yes / total_employees * 100

col1, col2, col3 = st.columns(3)

col1.metric("Tổng số nhân viên", total_employees)
col2.metric("Số nhân viên nghỉ việc", attrition_yes)
col3.metric("Attrition Rate", f"{attrition_rate:.2f}%")

# =========================
# 1. Pie chart Attrition Yes/No
# =========================
attrition_count = (
    df["Attrition"]
    .value_counts()
    .reset_index()
)

attrition_count.columns = ["Attrition", "Count"]

fig_pie = px.pie(
    attrition_count,
    names="Attrition",
    values="Count",
    title="Tỷ lệ Attrition Yes/No",
    hole=0.35
)

st.plotly_chart(fig_pie, use_container_width=True)

# =========================
# 2. Bar chart theo Department
# =========================
st.subheader("Attrition theo Department")

dept_attrition = (
    df.groupby(["Department", "Attrition"])
    .size()
    .reset_index(name="Count")
)

fig_dept = px.bar(
    dept_attrition,
    x="Department",
    y="Count",
    color="Attrition",
    barmode="group",
    title="Số lượng nhân viên nghỉ việc theo Department",
    text="Count"
)

fig_dept.update_traces(textposition="outside")

st.plotly_chart(fig_dept, use_container_width=True)

# =========================
# 3. Bar chart theo JobRole
# =========================
st.subheader("Attrition theo JobRole")

jobrole_attrition = (
    df.groupby(["JobRole", "Attrition"])
    .size()
    .reset_index(name="Count")
)

fig_jobrole = px.bar(
    jobrole_attrition,
    x="JobRole",
    y="Count",
    color="Attrition",
    barmode="group",
    title="Số lượng nhân viên nghỉ việc theo JobRole",
    text="Count"
)

fig_jobrole.update_layout(
    xaxis_tickangle=-45
)

fig_jobrole.update_traces(textposition="outside")

st.plotly_chart(fig_jobrole, use_container_width=True)
```
